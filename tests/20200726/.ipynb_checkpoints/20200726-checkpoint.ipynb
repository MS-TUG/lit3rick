```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
signalFile = "data/RAW_hilo1_450DAC.csv"
signal = np.genfromtxt(signalFile, delimiter=',') 
signal = signal - 2048
```

```python
N = len(signal)
f = [64.0* k / N for k in range(N)]
t = [128.0* k / N for k in range(N)]
fft = np.fft.fft(signal)
FFT = np.fft.fft(signal)

fft[0] = 0
for k in range(int(N/2 + 2)):
    if not (f[k] > 1.5 and f[k] < 5.0):
        fft[k] = 0
        fft[-k] = 0

filtered_signal = np.real(np.fft.ifft(fft))
```

```python
plt.figure(figsize = (15,5))
plt.subplot(1, 2, 2)

plt.plot(f[20:int(N/2)],np.abs(FFT[20:int(N/2)]),label="Raw FFT")
plt.plot(f[20:int(N/2)],np.abs(fft[20:int(N/2)]), label = "Filtered FT")
plt.legend()
plt.title("Frequencies")
plt.xlabel("MHz")
plt.subplot(1, 2, 1)
plt.plot(t,signal,alpha=0.5,label="raw")
plt.plot(t,filtered_signal,label="filtered")
plt.hlines(-2048,0,128,colors="grey",label="vMin")
plt.hlines(2047,0,128,colors="grey",label="vMax")
plt.legend()
plt.title("Signal and echoes")
plt.xlabel("us")
plt.savefig("example.jpg")
plt.show()
```
