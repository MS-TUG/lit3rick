```python
import numpy as np 
import matplotlib.pyplot as plt
import os
```

```python
allcsv = []
for file in os.listdir("./tmpdata/"):
    if file.startswith("k_"):
        filename = (os.path.join("./tmpdata/", file)) 
        signal = np.genfromtxt(filename, delimiter=',').astype(np.int16)
        print(file)
        allcsv.append(signal)
arr = np.array(allcsv).astype(np.int16)
print(arr)
```

```python
np.savez_compressed('all.npz', csv=arr)
```
