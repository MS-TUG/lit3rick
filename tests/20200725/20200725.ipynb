```python
import numpy as np 
import matplotlib.pyplot as plt
import os
```

```python
allcsv = []
for file in os.listdir("./CSVs/"):
    if file.startswith("k_"):
        filename = (os.path.join("./CSVs/", file)) 
        signal = np.genfromtxt(filename, delimiter=',') 
        plt.figure(figsize = (15,5))
        plt.plot(signal)
        
        plt.title(file)

        plt.savefig("imgs/"+file+".jpg")
        plt.show()
```
