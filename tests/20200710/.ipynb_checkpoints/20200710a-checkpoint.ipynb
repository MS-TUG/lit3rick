```python
import numpy as np
import matplotlib.pyplot as plt
from numpy import genfromtxt
import glob
```

```python
CSV3s = []
CSV0s = []
for x in glob.glob('./lit3/*hilo*.csv', recursive=True):
    CSV3s.append(x)
print(CSV3s)
for x in glob.glob('./un0/data/*.csv', recursive=True):
    CSV0s.append(x)
print(CSV0s)
```

```python
for filename in [x for x in CSV3s if "o" in x]:
    my_data = genfromtxt(filename, delimiter=',')
    if "4000" in filename:
        FN = filename.split("/")[-1].split(".")[0]
        L = len(my_data)
        Fech = 16
        t = [t/Fech for t in range(L)]
        f = [t*Fech/L for t in range(L)]
        my_data = my_data/2408.0 - 1
        plt.plot(t[0*Fech:138*Fech],my_data[0*Fech:138*Fech])
        signal3,t3,f3 = my_data[0*Fech:138*Fech],t[0*Fech:138*Fech],f
        
        plt.title("Details :"+FN)
        plt.savefig(FN+"_echo.jpg")
        plt.show()
```

```python
for filename in [x for x in CSV0s if "-2" in x]:
    my_data = genfromtxt(filename, delimiter=',')
    FN = filename.split("/")[-1].split(".")[0]
    L = len(my_data)
    Fech = 64
    t = [t/Fech+4.2 for t in range(L)]
    f = [t*Fech/L for t in range(L)]
    my_data = my_data/2
    plt.plot(t[0*Fech:138*Fech],my_data[0*Fech:138*Fech])
    signal0,t0,f0 = my_data[0*Fech:138*Fech],t[0*Fech:138*Fech],f
    plt.title("Details :"+FN)
    plt.savefig(FN+"_echo.jpg")
    plt.show()
```
