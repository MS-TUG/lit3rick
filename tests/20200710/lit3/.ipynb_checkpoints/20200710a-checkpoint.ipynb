```python
import numpy as np
import matplotlib.pyplot as plt
from numpy import genfromtxt
```

# Testing with shorted amps

```python
import glob
CSVs = []
for x in glob.glob('./*.csv', recursive=True):
    print(x)
    CSVs.append(x)
```

```python
for filename in CSVs:
    my_data = genfromtxt(filename, delimiter=',')
    L = len(my_data)
    Fech = 16
    t = [t/Fech for t in range(L)]
    f = [t*Fech/L for t in range(L)]

    plt.plot(t[0:320],my_data[0:320])
    plt.title(filename)
    plt.savefig(filename+"_start.jpg")
    plt.show()

    plt.plot(t[42*Fech:48*Fech],my_data[42*Fech:48*Fech])
    plt.title("Details :"+filename)
    plt.savefig(filename+"_echo.jpg")
    plt.show()

    fig= plt.figure(figsize=(15,5))
    plt.plot(t[0:2048+300],my_data[0:2048+300])
    plt.title("Overview :"+filename+ " - no oscilloscope probe")
    plt.savefig(filename+"_full.jpg")
    plt.show()

    FFT = np.abs(np.fft.fft(my_data))
    plt.plot(f[1000:int(L/2 - 1)],FFT[1000:int(L/2 - 1)])
    plt.title("FFT :"+filename)
    plt.savefig(filename+"_full.jpg")
    plt.show()
```
