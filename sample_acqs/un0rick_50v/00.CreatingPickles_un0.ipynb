```python
import pyUn0 as us

import numpy as np
import matplotlib.pyplot  as plt

from numpy import genfromtxt

try:
    import cPickle as pickle
except ImportError:  # Python 3.x
    import pickle
```

```python
x = us.us_json()
d={}
```

```python
plt.figure(figsize=(15,5))
for DAC in range(12):
    x.JSONprocessing("20201008a-"+str(DAC+1)+".json")
    if DAC == 11:
        DAC = "noise"
    else:
        DAC = int(x.tdac[2])
    d[DAC] = x.tmp[0:x.len_line]

    d["t"] = x.t[0:x.len_line]
    plt.plot(d["t"],d[DAC],label=str(DAC),alpha = 0.7)
plt.legend()
plt.title("Un0rick acquisitions")
plt.savefig("acqs_un0rick.jpg")
```

```python
with open('un0rick.pkl', 'wb') as fp:
    pickle.dump(d, fp, protocol=4)
```
